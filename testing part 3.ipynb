```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
# initialize projection and map limits/properties
map = Basemap(projection='hammer', lon_0=-80)
```

```python
#Draw package-included stuff
map.fillcontinents(color='#cc9966')

plt.show() #show plot
```

```python
#initialize projection and map limits/properties
map = Basemap(projection='ortho', lat_0=45, lon_0=-100, resolution='l')

#draw package-included stuff
map.drawcoastlines(linewidth=0.25)
map.drawcountries(linewidth=0.25)
map.fillcontinents(color='coral', lake_color='aqua')
map.drawmapboundary(fill_color='aqua')

plt.show() #show plot
```

```python
import numpy as np

#initialize projection and map limits/properties
map=Basemap(projection='mill', llcrnrlon=0, llcrnrlat=-80, urcrnrlon=360, urcrnrlat=80)

#draw package-included stuff
map.drawcoastlines(linewidth=0.25)
map.drawcountries(linewidth=0.25)
map.fillcontinents(color='coral',lake_color='aqua')
#draw edge of map projection region (the projection limb)
map.drawmapboundary(fill_color='aqua')
#draw lat/lon grid lines every 30 degrees
map.drawmeridians(np.arange(0,360,30))
map.drawparallels(np.arange(-90,90,30))

plt.show() #show plot
```

```python
#initialize projection and map limits/properties
map=Basemap(projection='mill', \
            llcrnrlon=-70, \
            llcrnrlat=42, \
            urcrnrlon=-50, \
            urcrnrlat=55, \
            resolution='l')

#draw package-included stuff
map.drawcoastlines(linewidth=0.25)
map.drawcountries(linewidth=0.25)
map.fillcontinents(color='gray', lake_color='#C2FFE0')
map.drawmapboundary(fill_color='#C2FFE0')

#make out data
lat, lon=44.45, -63.5 #halifax

#transform our data into projection-coordinates
x,y = map(lon, lat)

#draw our data
map.scatter(x,y,50, marker='o', color='r')

plt.show() #show plot
```
